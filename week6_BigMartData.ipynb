```python
import pandas as pd
```

```python
mart_train= pd.read_csv('Train.csv')
mart_train.head()
```

```python
mart_test= pd.read_csv('Test.csv')
mart_test.head()
```

```python
mart_data= pd.concat([mart_train, mart_test])
mart_data.head()
```

```python
mart_data.isnull().sum()
```

```python
mart_data.isnull().sum()
```

#### Label Encoding

```python
mart_data['Outlet_Location_Type_Encoded']  = mart_data['Outlet_Location_Type'] \
                                            .replace({'Tier 1': 1, 'Tier 2': 2, 'Tier 3': 3})
mart_data[['Outlet_Location_Type', 'Outlet_Location_Type_Encoded']].head()
```

#### One Hot Encoding for outlet_Type

```python
Outlet_Type_Dumm = pd.get_dummies(data=mart_data['Outlet_Type'], columns=['Outlet_Type'], 
drop_first=True)
pd.concat([mart_data['Outlet_Type'], Outlet_Type_Dumm], axis=1).head()
```

#### bin Item_MRP

```python
groups = ['Low', 'Med', 'High', 'Exp']
mart_data['Item_MRP_Bin_qcut'] = pd.qcut(mart_train['Item_MRP'], q=4, labels=groups)
mart_data[['Item_MRP', 'Item_MRP_Bin_qcut']].head()
```

#### frequency of this new variable

```python
mart_data['Item_MRP_Bin_qcut'].value_counts()
```

#### Alternatively

```python
#define bins 
bins = [0, 70, 140, 210, 280]
#name of groups
groups = ['Low', 'Med', 'High', 'Exp']
mart_data['Item_MRP_Bin_cut'] = pd.cut(mart_data['Item_MRP'], bins=bins, labels=groups)
mart_data[['Item_MRP', 'Item_MRP_Bin_cut']].head()
```

```python
mart_data['Item_MRP_Bin_cut'].value_counts()
```

#### Text Extraction

```python
mart_data['Item_Code'] = mart_data['Item_Identifier'].apply(lambda x: x[0:2])
mart_data[['Item_Identifier', 'Item_Code']].head()
```

#### Frequency Encoding

```python
# Frequency encoding using value_counts function 
Item_Type_freq = mart_data['Item_Type'].value_counts(normalize=True)
# Mapping the encoded values with original data 
mart_data['Item_Type_freq'] = mart_data['Item_Type'].apply(lambda x : Item_Type_freq[x])
print('The sum of Item_Type_freq variable:', sum(Item_Type_freq))
mart_data[['Item_Type', 'Item_Type_freq']].head(6)
```

#### Aggregation Features

```python
mart_data['Item_Outlet_Sales_Mean'] = mart_data.groupby(['Item_Identifier', 'Item_Type'])['Item_Outlet_Sales']\
                                     .transform(lambda x: x.mean())
mart_data[['Item_Identifier','Item_Type','Item_Outlet_Sales','Item_Outlet_Sales_Mean']].tail()
```
